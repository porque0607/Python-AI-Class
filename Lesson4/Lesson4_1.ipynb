```python
#學生總分為300分
#有些學生可加分5%
#如果加分超過300,就以300為止
try:
        scores = int(input("請輸入學生分數:(最高300分)"))
        is_add = input("學生是否符合加分條件?(Y,N)")

        if(is_add=='Y'):
            scores *= 1.05
            if scores >300:  #利用巢狀判斷增加條件
                scores = 300

        print(round(scores))  #將最後結果取四捨五入
        
except ValueError:
    print("輸入學生的分數,無法辨識")
except:
    print("出現錯誤,請重新執行")

print ("應用程式結束")
```
