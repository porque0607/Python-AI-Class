```python
#國文和數學分數
#如果同時100分,獎金1000元
#如果1科100分,獎金500元
#其他的0元
chinese = int(input("請輸入國文分數:"))
math = int(input("請輸入數學分數:"))

#邏輯運算子
bonus = 0
if chinese == 100 and math == 100:
    bonus = 1000
elif chinese == 100 or math == 100:
    bonus = 500
print(f"獎金{bonus}")
```
