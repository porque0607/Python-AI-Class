```python
#小明想存錢買機車,機車每輛30000元,將每個月的存款輸入,當足夠買機車時就顯示提示訊息告知

deposit = 0
while deposit < 30000:
   inputNum = int(input("請輸入本月份的存款:"))
   deposit += inputNum

print(f"恭喜!已經存夠了,存了{deposit}元")
```

```python
#小明想存錢買機車,機車每輛30000元,將每個月的存款輸入,當足夠買機車時就顯示提示訊息告知

deposit = 0
num = 0
while deposit < 30000:
    num+=1
    inputNum = int(input("請輸入本月份的存款:"))
    deposit += inputNum

print(f"恭喜!已經存夠了,存了{num}個月,共存了{deposit}元")
```

```python
#小明想存錢買機車,機車每輛30000元,將每個月的存款輸入,當足夠買機車時就顯示提示訊息告知

deposit = 0
num = 0
while True:
    num+=1
    inputNum = int(input("請輸入本月份的存款:"))
    deposit += inputNum
    if deposit >= 30000:
        break

print(f"恭喜!已經存夠了,存了{num}個月,共存了{deposit}元")
```

```python
#請設計一個程式,請使用者輸入數值,只有加總正偶數值,不加總正奇數值,如果輸入負數,結束程式

sum = 0
num = 0
while True:
    input_value = int(input(f"請輸入第{num+1}個數值"))
    if input_value < 0:
        break
    num+= 1
    if input_value % 2 == 1:
        continue
    sum += input_value
print(f"輸入的次數是{num},所有輸入的正偶數加總是:{sum}")
```
