```python
# 沒有參數,沒有傳出值的function
def sayHello():
    print("Hello!")

sayHello()
```

```python
def sayHello(name): #一個參數沒有傳出值
    print(f"哩賀!{name}!~")

sayHello("大美女")
```

```python
def sayHello(name): #一個參數,有傳出值
    return f"哩賀!{name}!~"

print(sayHello("大美女"))
```
