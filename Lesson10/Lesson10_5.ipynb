```python
#一般class 後面第一個字母都大寫
class Person:
    def __init__(self,name:str,age:int):   #自訂的初始化
     self.name = name
     self.age = age

#實體方法
    def description(self):
        print(f"我的名字叫{self.name}")
        print(f"我的年齡是{self.age}")
```

```python
p1 = Person(name="Robert",age=48)
p1.description()
```

```python
# Student繼承於Person
# Person是父類別
# Student是子類別
class Student(Person):
    pass
```

```python
s1 = Student(name="robert",age=40)
s1.description()
```
