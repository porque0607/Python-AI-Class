```python
#except 可以有好幾個,但只會執行一個會就跳出去
#Exception 表所有種類的error
#有好幾個except時,Exception要寫在最底下
# 處理錯誤
# 應用程式增加可靠度
try:
    money = int(input("請輸入金額:"))
    if money < 0:
        raise Exception("金額不可以小於0") 
except ValueError:
    print("格式錯誤")
except Exception as error:
    print(error)
else:
    print(money)
finally:
    print("應用程式結束")
```
