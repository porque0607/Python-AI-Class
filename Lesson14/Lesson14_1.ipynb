```python
import matplotlib.pyplot as plt

# 定義兩位學生各科分數（使用中文字作為科目名稱的鍵）
student1 = {"國文":78, "英文":80, "數學":92, "歷史":75, "探索":85}
student2 = {"國文":92, "英文":60, "數學":85, "歷史":62, "探索":73}

# 取得科目順序與對應的分數列表（保持同樣的科目順序以便比較）
subjects = list(student1.keys())     # ['國文', '英文', '數學', '歷史', '探索']
scores1 = list(student1.values())    # 學生1 的分數列表
scores2 = list(student2.values())    # 學生2 的分數列表

# 建立 Figure 與 Axes，設定解析度、大小與背景顏色
figure = plt.figure(dpi=100, figsize=(8,5), facecolor='lightgray')
axes = figure.add_subplot(1,1,1)  # 1x1 的子圖（即單一座標軸）

# 設定圖表標題與坐標軸標籤（中文字型在先前 cell 已設定）
axes.set_title("學生分數比較")
axes.set_xlabel("科目")
axes.set_ylabel("分數")

# 繪製兩條折線：學生1 與 學生2
# 使用 marker 標示每個點，並用不同的線條樣式以便分辨
axes.plot(subjects, scores1, marker='o', label='學生1', linestyle='--')
axes.plot(subjects, scores2, marker='s', label='學生2', linestyle='-.')

# 顯示圖例（會自動使用 label 內容）
axes.legend()

# 設定 x 軸範圍，讓第一與最後一個科目點不貼齊邊界（-0.5 到 len-0.5）
axes.set_xlim(-0.5, len(subjects)-0.5)

# 計算 y 軸刻度：從兩組分數的最小值往下延伸 10（避免貼齊），到最大值往上延伸 10，
# 並以 10 為刻度間隔（range 需整數）
y_min = min(min(scores1), min(scores2)) - 10
y_max = max(max(scores1), max(scores2)) + 10
axes.set_yticks(range(y_min, y_max, 10))

# 在每個資料點上方加上數值標示
# enumerate 回傳索引 i 與對應的分數 score；使用科目字串作為 x 座標並顯示分數文字
for i, score in enumerate(scores1):
    axes.text(subjects[i], score + 1, str(score), ha='center', va='bottom')  # 學生1 的分數標示
for i, score in enumerate(scores2):
    axes.text(subjects[i], score + 1, str(score), ha='center', va='bottom')  # 學生2 的分數標示

# 只顯示 y 軸的格線（便於比較分數高低）
axes.grid(axis='y')

# 顯示圖表
plt.show()
```
